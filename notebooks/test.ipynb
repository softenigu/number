```python
def grade(score):
    if score >= 80:
        return "A"
    elif score >= 60:
        return "B"
    else:
        return "C"

students = [
    ("Aki", 92),
    ("Ken", 67),
    ("Mio", 58),
]

for name, score in students:
    g = grade(score)
    print(f"{name}  score={score}  grade={g}")
```

〇　サイコロの目で６が出る可能性

```python
import numpy as np
import matplotlib.pyplot as plt

# np.random.seed(0)

n = 50
rolls = np.random.randint(1, 7, size=n)
p_six = (rolls == 6).cumsum() / np.arange(1, n+1)

plt.figure(figsize=(7,4))
plt.plot(p_six)
plt.axhline(1/6, linestyle="--")
plt.title("P(roll=6) converges as we roll more dice")
plt.xlabel("Number of rolls")
plt.ylabel("Estimated probability")
plt.show()
```
